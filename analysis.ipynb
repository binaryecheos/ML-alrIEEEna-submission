```python
import numpy as np
import pandas as pd
import sklearn
```

```python
X_test = pd.read_csv(r"TEST.csv")
X_test = X_test.drop("ID", axis=1)
df_train = pd.read_csv(r"TRAIN.csv")
```

```python
df_train
```

```python
X_train = df_train.drop("Class", axis=1)
X_train
```

```python
y_train = df_train["Class"]
y_train
```

```python
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()
X_train = scaler.fit_transform(X_train)
X_test = scaler.fit_transform(X_test)

```

```python
X_test.shape
```

```python
skew_values = X=df_train.skew(numeric_only=True)
print(skew_values)
```

```python
X_trans = df_train.drop('Class', axis=1)
```

```python
from sklearn.preprocessing import PowerTransformer

pt = PowerTransformer(method="yeo-johnson")

skewed_cols = X_trans.skew(numeric_only=True)
skewed_cols = skewed_cols[abs(skewed_cols) > 1].index
df_train[skewed_cols] = pt.fit_transform(df_train[skewed_cols])
```

```python
X_train = df_train.drop("Class", axis=1)
y_train = df_train["Class"]
```

```python
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score, f1_score

model = LogisticRegression(max_iter=1000)

model.fit(X_train, y_train)

y_pred = model.predict(X_test)
```

```python
y_pred
```

```python
from sklearn.model_selection import cross_val_score

scores = cross_val_score(model, X_train, y_train, cv=5, scoring="f1")

print(scores)
print("Mean F1:", scores.mean())
```

```python
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import accuracy_score, f1_score

model = RandomForestClassifier(
    n_estimators=100,
    random_state=42
)

model.fit(X_train, y_train)

y_pred = model.predict(X_test)
```

```python
from sklearn.model_selection import cross_val_score

scores = cross_val_score(model, X_train, y_train, cv=5, scoring="f1")

print(scores)
print("Mean F1:", scores.mean())
```
